```python
from PyALE import ale
import numpy as np
import joblib
import matplotlib.pyplot as plt
import pandas as pd
import warnings
warnings.filterwarnings("ignore")
```

```python
def plot_ale(idx):
    fig,ax = plt.subplots(1, figsize=(15, 7))
    ax.set_xlabel('ALE-%s'%var_name[idx])
    ale_eff = ale(X=df,model  = model,feature = [name[idx]],grid_size=50,include_CI=False,feature_type='continuous',fig = fig,ax = ax)
    plt.savefig('../output/month_mean/ale%s'%name[idx])
```

```python

x_train = np.load("../data/mm_balanced.npy")
echoes = x_train[:,[6,14,16,27,34,35,37]].copy()
name = ['pftCrop','GDP','Distance_to_populated_areas','WDPA_fracCover','road_density','Livestock','pop_density']
var_name = ['CropLand', 'GDP', 'Distance to population',
        'Protected areas', 'Road Density', 'Livestock', 'Population Density']
x_train = np.delete(x_train,[0,1,2,15,18,19,21,22,23,24,25,26,32,36],axis = 1)
df = pd.DataFrame(x_train,columns=['agb', 'pft_fracCover', 'sm', 'pftCrop',
       'pftHerb', 'pftShrubBD', 'pftShrubNE', 'pftTreeBD', 'pftTreeBE',
       'pftTreeND', 'pftTreeNE ', 'GDP', 'Distance_to_populated_areas',
       'fPAR', 'vod_K_anomalies', 'WDPA_fracCover', 'dtr', 'pet', 'tmx', 'wet', 'precip',
       'Livestock', 'road_density', 'pop_density'])
model = joblib.load('../output/month_mean/feature_all.joblib')
for i in range(echoes.shape[1]):
    plot_ale(i)
```
