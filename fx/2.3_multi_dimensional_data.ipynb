#### Data dimensionality

A dimension is defined by the Oxford dictionary as
- a measurement of something in a particular direction, especially its height, length, or width
- a part or feature or way of considering something

The first definition is just a special case of the second. 
For geoscience data, it is important to **think about dimensions as features** describing data, including but beyond spatial coordinates

For example, microwave remote sensing measurements might be characterized by 6 features, i.e., dimensions:
- time of observation
- latitude
- longitude
- microwave frequency
- wave polarization
- antenna incidence angle

#### n-dimensional data in python

Recall that pandas is built around 1D or 2D numpy arrays to provide useful labels to rows and columns.

Similarly, [xarray](https://docs.xarray.dev/en/stable/) is built around n-dimensional numpy arrays and particularly well suited to work with netCDF files

![](./figures/2.1_xarray.png)

```python
import warnings
warnings.filterwarnings("ignore")

%matplotlib inline
import matplotlib.pyplot as plt
import cartopy.crs as ccrs

import numpy as np
import pandas as pd
import xarray as xr

from netCDF4 import Dataset
```

```python
# Data dimensions / features
dates = pd.date_range('2000-01-01', '2000-12-31', freq='D')
lats = np.arange(-89.5, 90, 1)
lons = np.arange(-179.5, 180, 1)
colors = ['blue', 'orange', 'green']

# Creating some random data with the extent matching the dimensions we have defined above
data = np.random.rand(len(dates), len(lats), len(lons), len(colors))
coordinates = [dates, lats, lons, colors]
dimensions = ['time', 'latitude', 'longitude', 'color']

# Creating a xarray DataArray with the dimensions defined above 
darr = xr.DataArray(data, 
                    coords=coordinates, 
                    dims=dimensions)
```

The DataArray is the 'fundamental unit' of xarray, comparable to a pandas Series 

But: instead of the pandas 'index' (in our case, the time stamps), we now have multuple dimensions discriminating data points, referred to as 'coordinates'.

```python
# Coordinates are stored as an attribute
print(darr.coords, '\n')

# The underlying 'data' is again a numpy array.
print(darr.data.shape)
```

Similar to the pandas Dataframe, which holds multiple columns with the same index, 
xarray offers the 'Dataset', which contains multiple so-called 'variables' all of which have the same 'coordinates' (i.e, 'dimensions')

**Note:** Pandas should really better be considered to contain only 1-dimensional data, the dimension being the 'Index', either a single 1D data set within a pd.Series, or multiple 1D data sets within a pd.Dataframe. xarray contains either single (DataArray) or multiple (Dataset) n-dimensional data sets.

```python
# Creating some more variables
data2 = np.random.rand(len(dates), len(lats), len(lons), len(colors)) 
data3 = np.random.rand(len(dates), len(lats), len(lons), len(colors))


# Create a Dataset from these variables which all have the same dimensions as the DataArray before
ds = xr.Dataset({'var1': (dimensions, data),
                 'var2': (dimensions, data2),
                 'var3': (dimensions, data3)},
                coords=dict(zip(dimensions,coordinates)))
```

```python
print(ds.coords, '\n')

print(ds.data_vars, '\n')
```

```python
# Just like each column of a pandas Dataframe is actually a pandas Series, 
# each Dataset variable is a DataArray:

print(type(ds), '\n')

print(type(ds['var1']))
```

#### xarray querying

Remember that xarray is essentially a wrapper around n-dimensional numpy arrays that assigns labels to each dimension

This serves the purpose of more intuitive / user-friendly data access than the typical integer indexing of (numpy) arrays 

```python
# DataArray indexing by label
darr.loc['2000-02-01':'2000-02-15', 0:3, 0:3, 'blue']
```

```python
# DataArray indexing by label including dimension names
darr.sel(time=slice('2000-02-01','2000-02-15'), 
         latitude=slice(0,3),
         color='blue')
```

```python
# DataArray indexing by integer including dimension names
darr.isel(time=slice(0,10), color=-1)
```

Note that the **sel/isel** method are **functions with positional arguments**, which is slightly more convenient when slicing along only a single dimension.

```python
# Dataset querying is identical to DataArray indexing using sel and isel
ds.sel(time=slice('2000-02-01','2000-02-15'), 
       latitude=slice(0,3),
       color='orange')
```

```python
# If a DataArray is sliced down to just one or two features, 
# it can be converted back into a pandas Series or Dataframe
print(darr.isel(latitude=0, longitude=0).to_pandas().head(5))
```

#### Special time series functionality

The 'time' dimension holds a special place among attributes in that it allows for very convenient [date operations](https://docs.xarray.dev/en/stable/user-guide/time-series.html#) similar to pandas

The prerequisite is that the time dimension holds time stamps as datetime64 data, which can be ensured using pandas.

```python
# Createing single time steps
print(pd.to_datetime(["2000-01-01", "2000-02-02"]), '\n')

# Createing time slices
print(pd.date_range("2000-01-01", periods=365), '\n')
```

```python
# Datetime attributes for easy querying can be accessed using a special .dt accessor
darr[darr.time.dt.dayofweek == 2]
```

```python
# .dt can equally be used for slicing:
ds.isel(time=(ds.time.dt.month == 2))
```

#### Data manipulation

xarray also offers a plethora of functionality to be performed on Datasets or DataArray.

These include [data interpolation](https://xarray.pydata.org/en/stable/user-guide/interpolation.html), [array math operations](https://xarray.pydata.org/en/stable/user-guide/computation.html), and many others (see the [full documentation]()).

The syntax is very similar to those we used for pandas.

```python
# e.g.: Averaging across multiple dimensions
darr.mean(dim=['time','color'])
```

Notice how the result remains a DataArray, which would allow for easy visualization (see the next section).

#### Data visualization

xarray is also well integrated with matplotlib

Datasets can be readily plotted when queried correctly

```python
# Create some differences between random data in the feature dimensions
darr.loc[:,:,:,'orange'] += 1
darr.loc[:,:,:,'blue'] += 2
```

```python
# Plotting can account for two features
darr.sel(latitude=23.5, longitude=164.5).plot.line(x="time", figsize=(15,4))
```

```python
# Depending on the data dimensions, DataArray.plot() figures out whether to create a 1D or a 2D plot.
darr.var(dim=['time','color']).plot()
```

#### netCDF integration

xarray is particularly well integrated with netCDF files.

The Network Common Data Format, [NetCDF](https://www.unidata.ucar.edu/software/netcdf/), is the community standard for array-like scientific data. 

It is a container format that is - not surprisingly - structured very similarly to xarray Datasets (or rather, xarray Datasets have been designed to resemble the NetCDF file structure).

NetCDF files can contain different variables, which all are allowed to have an arbitrary number of dimensions.

```python
# Datasets can be written out directly as netCDF files 
darr.to_netcdf('./Data/ds.nc')
```

```python
# netCDF files can be read directly into a Dataset
ds = xr.open_dataset('./Data/2_example.nc')
print(ds.coords, '\n')
print(ds.data_vars, '\n')
print(ds['SM'].shape)
```

```python
# Visualizing in the spatial dimension
plt.figure(figsize=(8,8))
ax = plt.subplot(projection=ccrs.PlateCarree())
ax.coastlines()
ds['VOD'].mean(dim='time').plot(x='lon', y='lat',
                                ax=ax,
                                vmin=0, vmax=0.8,
                                cmap='YlGn')
```

```python
# Visualizing in the temporal dimension
plt.figure(figsize=(16,5))
ds['VOD'].isel(lat=120, lon=60).dropna(dim='time').plot()
```

#### Attributes and CF conventions

netCDF files can store metadata that describe the different variables or dimensions, referred to as attributes.

The [CF conventions](https://cfconventions.org) are community standards that "define metadata that provide a definitive description of what the data in each variable represents, and the spatial and temporal properties of the data."

xarray can readily understand and manipulate attributes. 

```python
# Attributes are particularly useful to define more meaningful coordinate/variable names, units, etc.
print(ds['lat'].attrs, '\n')

print(ds['SM'].attrs, '\n')
```

**NOTE:** Many software packages have special built-in functionalities that depend upon the CF conventions, such as recognizing spatial dimension by standardized names (e.g., time, latitude, longitude).

xarray, for example automatically parses some attributes without notice:

```python
# Read the same netCDF file with the netCDF4 package instead of xarray
nc = Dataset('./Data/2_example.nc')
```

```python
# xarray-parsed attributes
print(ds['time'].attrs, '\n')

# original netCDF4 file attributes
print(nc['time'].__dict__)
```

```python
# xarray-parsed attributes
print(ds['SM'].attrs, '\n')

# original netCDF4 file attributes
print(nc['SM'].__dict__)
```

Attributes are stored globally as well as for each variable / coordinate separately.

Attributes can be set in xarray when creating a DataArray or Dataset, but care must be taken to assign them correctly globally or to the correct variable or dimension.

Alternatively, attributes can be set for an existing xarray object through a dictionary:

```python

ds['SM'].attrs = {'long_name': 'volumetric soil moisture content',
                  'units': 'm3 m-3'}
```
