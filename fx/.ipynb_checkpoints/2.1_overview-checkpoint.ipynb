#### Pandas data structure for time series

```python
import warnings
warnings.filterwarnings("ignore")

%matplotlib inline
import matplotlib.pyplot as plt
import cartopy.crs as ccrs

import numpy as np
import pandas as pd
import xarray as xr
```

```python
# Read example time series data from a text file
ts = pd.read_csv('./Data/2_example.csv', index_col=0, parse_dates=True)
print(ts.head(10))
```

```python
# Plot some columns
fig, ax = plt.subplots(3,1, sharex=True, figsize=(16,10))

ts['soil temperature'].plot(ax=ax[0])
ax[0].set_title('Soil temperature');
ax[0].axhline(color='black', linestyle='--')

ts['snow mass'].plot(ax=ax[1])
ax[1].set_title('Snow mass');

# Perform some data manipulations
ts['soil temperature'].resample('30D').mean().plot(ax=ax[2])
ax[2].set_title('Monthly mean temperature');
```

#### xarray data structures for multi-dimensional data

```python
# netCDF files can be read directly into a Dataset
ds = xr.open_dataset('./Data/2_example.nc')
print(ds.coords, '\n')
print(ds.data_vars, '\n')
```

```python
# Visualizing in the temporal dimension
ds['VOD'].isel(lat=120, lon=60).dropna(dim='time').plot(figsize=(16,4));
```

```python
# Visualizing in the spatial dimension
plt.figure(figsize=(8,8))
ax = plt.subplot(projection=ccrs.PlateCarree())
ds['VOD'].mean(dim='time').plot(vmin=0, vmax=0.8,
                                cmap='YlGn')
ax.coastlines();
```
